```python
import pandas as pd
```

```python
df = pd.read_csv("Cleaned_Shipping_Details.csv")
df.drop(columns=df.columns[0], axis=1,  inplace=True)
df
```

```python
df.info()
```

```python
# extract specific columns by name,
arm_df = df[["Material", "Transport", "Fragile","Remote Location","Customer Information","Installation Included","International","Express Shipment","state"]]
```

```python
arm_df
```

```python
arm_df["Fragile"].value_counts()
```

```python
# Dictionaries to replace Yes & No with appropriate descriptions
replace_dict_Fragile = {'Yes': 'Fragile', 'No':'Not_Fragile'}
replace_dict_Remote_Location = {'Yes': 'Remote_Location', 'No':'Not_Remote_Location'}
replace_dict_Installation_Included = {'Yes': 'Installation_Included', 'No':'Installation_Not_Included'}
replace_dict_International = {'Yes': 'International_Shipment', 'No':'Not_International_Shipment'}
replace_dict_Express_Shipment = {'Yes': 'Express_Shipment', 'No':'Not_Express_Shipment'}


arm_df["Fragile"] = arm_df["Fragile"].replace(replace_dict_Fragile)
arm_df["Remote Location"] = arm_df["Remote Location"].replace(replace_dict_Remote_Location)
arm_df["Installation Included"] = arm_df["Installation Included"].replace(replace_dict_Installation_Included)
arm_df["International"] = arm_df["International"].replace(replace_dict_International)
arm_df["Express Shipment"] = arm_df["Express Shipment"].replace(replace_dict_Express_Shipment)
```

```python
arm_df
```

```python
transactions_list = arm_df.values.tolist()
```

```python
transactions_list
```

```python
import csv

# Save transaction list as CSV file
with open('transactions.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(transactions_list)
```
