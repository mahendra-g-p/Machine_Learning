```python
import requests

url = "https://ebay-search-result.p.rapidapi.com/search/ipad"

headers = {
	"X-RapidAPI-Key": "52a8d64c74mshe737ebd1b1e1f33p1e57a6jsn441d8c34764b",
	"X-RapidAPI-Host": "ebay-search-result.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers)

#print(response.text)
```

```python
jsontxt = response.json()
print(jsontxt) 
```

```python
MyFILE=open(filename, "a")
#LABEL=topic
for items in jsontxt["results"]:
    Item_Title=items["title"]
    Item_Title=str(Item_Title)
    Item_Title=Item_Title.replace(',', '')
    
        
    Item_Price=items["price"]
    Item_Price = str(Item_Price)
    #Item_Price = Item_Price.replace(',','')
    Item_Price=re.sub(r'[,;@#?!&$\-\']+', '', str(Item_Price), flags=re.IGNORECASE)
    
        
    Shipping_Cost=items["shipping"]
    Shipping_Cost = str(Shipping_Cost)
    #Item_Price = Item_Price.replace(',','')
    Shipping_Cost=re.sub(r'[,;@#?!&$\-\']+', ' ', str(Shipping_Cost), flags=re.IGNORECASE)

    
    Shipping_Cost = Shipping_Cost.replace('+','')
    Split_data_1 = Shipping_Cost.split()
    if len(Split_data_1)>0:
        Shipping_Price = Split_data_1[0]
    else:
        Shipping_Price = "0.00" 
        
    location_text=items["location"]
    location_text_split = location_text.split()
    Shipping_location=" ".join(location_text_split[1:])
        
    Rating_text=items["rating"]
    Rating_text_split = Rating_text.split()
    if len(Rating_text_split)>0:
        Rating = Rating_text_split[0]
    else:
        Rating = "NA"
        
    WriteThis=str(Item_Title)+","+str(Item_Price)+","+str(Shipping_Price)+","+ str(Shipping_location) + "," + str(Rating) + "\n"
    #print(WriteThis)
        
    MyFILE.write(WriteThis)
        
        
MyFILE.close()
```
